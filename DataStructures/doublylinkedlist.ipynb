```python
class Node:
    def __init__(self,data=None,next=None,prev=None):
        self.data = data
        self.next = next
        self.prev = prev
    
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_start(self,data):
        if self.head is None:
            self.head = Node(data,None,None)
            return
        else:
            node = Node(data,self.head,None)
            self.head.prev = node
            self.head = node
            return
    
    def print_forward(self):
        if self.head is None:
            print("Empty list")
        
        else:
            itr = self.head
            flist = ""
            while itr:
                flist += str(itr.data) + "-->"
                itr =  itr.next
            print(flist)
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count+=1
            itr = itr.next
        return count

    def get_lastnode(self):
        itr = self.head
        while itr.next:
            itr = itr.next
        return itr    

    def print_backard(self):
        itr = self.get_lastnode()
        backll = ""
        while itr:
            backll += str(itr.data) + "-->"
            itr = itr.prev
        print(backll)

    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data,None,None)
            return
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data,None,itr)
    
    def insert_list(self,data):
        self.head = None
        for i in data:
            self.insert_at_end(i)
    
    def insert_at(self,index,data):
        if index < 0 or index > self.get_length():
            raise Exception("Out of bound")
        
        if index == 0:
            self.insert_at_start(data)
            return
        itr = self.head
        count = 0
        while count < index-1:
            itr = itr.next
            count+=1
        node = Node(data,itr.next,itr)
        itr.next.prev = node
        itr.next = node


    def remove_at(self,index):
        if index < 0 or index >= self.get_length():
            raise Exception("Out of bound")

        if index == 0:
            self.head = self.head.next
            self.head.prev = None
            return
        itr = self.head
        count = 0
        while itr:
            if count == index:
                itr.prev.next = itr.next
                while itr.next:
                    itr.next.prev = itr.prev
                    break
            count +=1
            itr = itr.next
        
        

if __name__ =="__main__":
    dd = DoublyLinkedList()
    dd.insert_list([0,1,2,3,4,5])
    dd.print_forward()
    
    dd.remove_at(5)
    dd.print_forward()
    #dd.print_backard()
    #print(dd.get_lastnode().data)
    
```
