# Queue

```python
from collections import deque
class Queue:
    def __init__(self):
        self.buffer = deque()

    def enqueue(self,data):
        self.buffer.appendleft(data)

    def deque(self):
        if len(self.buffer):
            return self.buffer.pop()
        return False
    
    def len(self):
        return len(self.buffer)
    
    def isempty(self):
        return len(self.buffer) == 0
    
if __name__ == "__main__":
    queue = Queue()
    queue.enqueue(2)
    queue.enqueue(3)
    queue.enqueue(4)

    print(queue.buffer)

    queue.deque()
    queue.deque()
    print(queue.len())
    print(queue.buffer)
```

```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_start(self,data):
        node = Node(data,self.head)
        self.head = node

    def print_the_list(self):
        if self.head is None:
            return "List is empty"
        ans  = ""
        itr = self.head

        while itr:
            ans += str(itr.data) + "--->"
            itr  = itr.next
        print(ans)
    
    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data,None)
            return
        itr = self.head
        while itr.next:
            itr =  itr.next
        itr.next = Node(data,None)
    
    def insert_list(self,data):
        self.head = None

        for i in data:
            self.insert_at_end(i)
    
    def length(self):
        count = 0
        itr = self.head
        while itr:
            count +=1
            itr = itr.next
        return count

    def insert_at(self,index,data):
        if index == 0:
           self.head= Node(data,self.head)
           
           return
        count = 0 
        itr = self.head
        while count < index-1:
            count +=1
            itr = itr.next
        newnode = Node(data,itr.next)
        itr.next = newnode
        return
    
    def insert_after_value(self,value,data):
        itr = self.head
        while itr.data != value:
            itr = itr.next
        itr.next = Node(data,itr.next)
        return
    
    def remove_at(self,index):
        if index == 0:
            self.head = self.head.next
            return
        count = 0
        itr = self.head
        while count < index-1:
            count +=1
            itr = itr.next
        itr.next = itr.next.next
        return
    
    def remove_by_value(self,value):
        if self.head is None:
            return
        if self.head.data == value:
            self.head = self.head.next
            return
        itr = self.head
        while itr.next:
            
            if itr.next.data == value:
                itr.next = itr.next.next
                return
            itr = itr.next
        
    def reverse(self):
        prev,curr =None,self.head
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        self.head = prev
        return self.print_the_list()

    

        
    
if __name__ == "__main__":
    ll = LinkedList()
    ll.insert_list([0,1,2,3,4])
    ll.print_the_list()
    ll.reverse()
```
